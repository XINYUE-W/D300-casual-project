# Delgado Context

This notebook shows a Delgado-style Engel curve background figure: food share against log expenditure, with a GWL-style fit and a LOWESS smooth.

## Outputs Produced By This Notebook

Running this notebook writes the Delgado background figure to `outputs/context_results/`:
- `engel_curve_gwl_vs_lowess.png`


```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from pathlib import Path
import importlib.util

_cwd = Path.cwd().resolve()
_cands = [_cwd, _cwd.parent]
PROJECT_ROOT = next((p for p in _cands if (p / "MODELS" / "01_engel_curve_context.py").exists()), None)
if PROJECT_ROOT is None:
    raise FileNotFoundError("Could not locate project root")

MODELS_DIR = PROJECT_ROOT / "MODELS"
OUTPUT_DIR = PROJECT_ROOT / "outputs" / "context_results"

def load_module(module_filename: str, module_name: str):
    module_path = MODELS_DIR / module_filename
    spec = importlib.util.spec_from_file_location(module_name, module_path)
    if spec is None or spec.loader is None:
        raise ImportError(f"Could not load {module_path}")
    module = importlib.util.module_from_spec(spec)
    spec.loader.exec_module(module)
    return module

context_module = load_module("01_engel_curve_context.py", "engel_curve_context")
```

```python
context_module.main()
```
